**2. La funci√≥n `describir(datos)` que generalice la funci√≥n `generar_descripcion` para recibir una lista de frutas y devolver una lista con las descripciones de cada una:**

Por ejemplo, `describir([limon, naranja]])` deber√° devolver una lista de la forma `[descripcion_limon, descripcion_naranja]`

````python
def describir(datos):
    """Devuelve la descripci√≥n de cada fruta de una lista de frutas en base a su nombre, tipo, color y si presenta pepas o no.

    Parameters
    ----------
    datos : list of dict
        Una lista en donde cada uno de sus elementos corresponde a un diccionario que posee las siguientes llaves:
        - "nombre" : str, nombre de la fruta
        - "tipo" : str, tipo de fruta
        - "color" : str, color de la fruta
        - "pepas" : bool, si es True indica que la fruta s√≠ presenta pepas. Si es False, la fruta no tiene pepas.


    Returns
    -------
    list of str
        Lista de cadenas de caracteres en donde cada una posee la descripci√≥n de una fruta en base a sus caracter√≠sticas
        La descripci√≥n de cada fruta incluye su respectivo art√≠culo (El/La), indica el nombre de la fruta, su tipo, color y si presenta pepas o no
    """

    descripciones = []
    for i in range(len(datos)):
        descripciones.append(generar_descripcion(datos[i]))
    return descripciones
````

**3. La funci√≥n `filtrar_por_pepa(datos, tiene_pepas)` que dado un tipo de fruta y un booleano, retorne las frutas que tienen o no tienen pepas (seg√∫n el valor de `tiene_pepas`).**
    

Por ejemplo, `filtrar_por_pepa(datos, True)` deber√° retornar una lista con los diccionarios de: naranja, limon, frutilla y frambuesa.

````python
def filtrar_por_pepa(datos, tiene_pepas):
    """Dada una lista de frutas, retorna aquellas que poseen pepas o no, de acuerdo a lo indicado por tiene_pepas

    Parameters
    ----------
    datos : list of dict
        Una lista en donde cada uno de sus elementos corresponde a un diccionario que posee las siguientes llaves:
        - "nombre" : str, nombre de la fruta
        - "tipo" : str, tipo de fruta
        - "color" : str, color de la fruta
        - "pepas" : bool, si es True indica que la fruta s√≠ presenta pepas. Si es False, la fruta no tiene pepas.
    tiene_pepas : bool
        Si es True, se retornan las frutas que s√≠ tienen pepas. Si es False, se retornan las frutas que no tienen pepas.

    Returns
    -------
    list of dict
        Lista en donde cada uno de sus elementos correponde a un diccionario de una fruta
        Solo posee las frutas que fueron filtradas, seg√∫n se haya indicado, con o sin pepas.
    """

    filtrados = []
    for i in range(len(datos)):
        if (datos[i]['pepas'] == tiene_pepas):
            filtrados.append(datos[i])
    return filtrados
````

**4. La funci√≥n `conteo_colores(datos)` que cree un diccionario que haga un conteo los colores de las frutas.**

Para estos datos, la funci√≥n deber√≠a retornar el siguiente diccionario: 
    
```python
{"amarillo": 3, "naranjo": 1, "rojo": 2}
```

````python
def conteo_colores(datos):
    """Devuelve un diccionario que posee el conteo, por color, de las frutas en datos.

    Parameters
    ----------
    datos : list of dict
        Una lista en donde cada uno de sus elementos corresponde a un diccionario que posee las siguientes llaves:
        - "nombre" : str, nombre de la fruta
        - "tipo" : str, tipo de fruta
        - "color" : str, color de la fruta
        - "pepas" : bool, si es True indica que la fruta s√≠ presenta pepas. Si es False, la fruta no tiene pepas.


    Returns
    -------
    dict
        Diccionario que posee el n√∫mero de frutas por color. Las llaves corresponden a los respectivos colores.
    """
    conteo = {}
    for i in range(len(datos)):
        if datos[i]["color"] not in conteo:
            conteo[datos[i]["color"]] = 1
        else:
            conteo[datos[i]["color"]] += 1
    return conteo
````

**5. La funci√≥n `obtener_tipos` que devuelva una lista con los tipos √∫nicos de fruta usando `set()`**.

Por ejemplo, `obtener_tipos([limon, naranja, platano])` deber√≠a devolver `["citrica", "tropical"]`

````python
def obtener_tipos(datos):
    """Devuelve una lista con los tipos √∫nicos de las frutas en datos.

    Parameters
    ----------
    datos : list of dict
        Una lista en donde cada uno de sus elementos corresponde a un diccionario que posee las siguientes llaves:
        - "nombre" : str, nombre de la fruta
        - "tipo" : str, tipo de fruta
        - "color" : str, color de la fruta
        - "pepas" : bool, si es True indica que la fruta s√≠ presenta pepas. Si es False, la fruta no tiene pepas.


    Returns
    -------
    list of str
        Lista en donde cada uno de sus elementos es un string que corresponde a un tipo de fruta.
        Cada elemento de la lista es √∫nico, representan los tipos √∫nicos de las frutas en datos.
    """
    tipos = set()
    for i in range(len(datos)):
        tipos.add(datos[i]["tipo"])
    tipos = list(tipos)
    return tipos
````

- Un m√©todo llamado `esta_enchufado(self)` que levante una excepci√≥n `Exception` y que termine con la ejecuci√≥n del programa cuando el atributo enchufado sea `False`. La excepci√≥n debe levantar el mensaje `'Alerta ‚ö†Ô∏è: El electrodom√©stico no est√° enchufado'`
- Un m√©todo llamado `enchufar(self)` que cambia el estado de `enchufado` a True.



````python
class Electrodomestico:
    """
    Clase que representa un electrodom√©stico

    ...

    Atributos
    ----------
    enchufado: bool
      Indica si el electrodom√©stico est√° enchufado.

    M√©todos
    -------
    esta_enchufado()
      Si enchufado es False, levanta una excepci√≥n.
    enchufar()
      Cambia el estado de enchufado de False a True.

    """

    def __init__(self):
      self.enchufado = False
        
    def esta_enchufado(self):
      if not self.enchufado:
         raise Exception(
            'Alerta ‚ö†Ô∏è: El electrodom√©stico no est√° enchufado'
         )
            
    def enchufar(self):
      self.enchufado = True
````

````python
class Jugera(Electrodomestico):
    """
    Representa el electrodom√©stico Juguera

    ...

    Atributos
    ---------
    bandeja: list
        Caja para ingredientes
    
    M√©todos
    -------
    agregar_ingrediente(nueva_fruta)
        Agrega una nueva fruta a la bandeja
    listar_ingredientes()
        Imprime un mensaje con las frutas presentes en la bandeja,
        o "Bandeja vac√≠a" si no hay ninguna
    bandeja_vacia()
        Si la bandeja est√° vac√≠a levanta una excepci√≥n
    preparar_jugo()
        Prepara un jugo con los ingredientes de la bandeja, vaci√°ndola.
        Verifica que todo est√© en orden antes de hacerlo, y de no estarlo, levanta excepciones.
    
    """
    
    def __init__(self):
      super().__init__()
      self.bandeja = []
        
    def agregar_ingrediente(self, nueva_fruta):
        self.bandeja.append(nueva_fruta)
     
    def listar_ingredientes(self):
        if not self.bandeja:
            print("Bandeja vac√≠a")
        else:
            print(f"Ingredientes en la bandeja: {', '.join((fruit['nombre'] for fruit in self.bandeja))}")
    
    def bandeja_vacia(self):
        if not self.bandeja:
            raise Exception(
                'Error ‚ùå: La bandeja no tiene ingredientes.'
            )
    
    def preparar_jugo(self):
        try:
            self.esta_enchufado()
            self.bandeja_vacia()
            if filtrar_por_pepa(self.bandeja, True):
                print('Alerta ‚ö†Ô∏è: El jugo puede contener restos de pepas.')
            mensaje = f"Jugo de {', '.join((fruit['nombre'] for fruit in self.bandeja))} listo. üèñÔ∏èü•§ Que lo disfrutes!!! ü•§üèñÔ∏è. "
            self.bandeja = []
        except Exception as e:
            mensaje = e
        return mensaje
````

### Clase Jugera + Properties  [1.0 puntos]

Implementar `bandeja` usando una `property` que permita setear una `nueva_bandeja` como `bandeja` seg√∫n las siguientes condiciones:


- Compruebe que `nueva_bandeja` sea una lista. En caso contrario, levante una excepci√≥n.
- No permita agregar m√°s de 3 ingredientes a la bandeja a la vez. Si se entregan m√°s de 3 frutas, se levante una excepci√≥n.
- Se compruebe que todos los elementos de la lista sean frutas. Para esto, por cada fruta compruebe que:
    1. La fruta sea diccionario.
    2. El diccionario entregado tenga las llaves `nombre`, `tipo`, `color` y `pepas`.

````python
class Jugera(Electrodomestico):
    """
    Representa el electrodom√©stico Juguera

    ...

    Atributos
    ---------
    _bandeja: list
        Caja para ingredientes
    
    M√©todos
    -------
    bandeja()
        M√©todo getter para acceder al atributo _bandeja
    bandeja.setter()
        M√©todo setter para cambiar el atributo _bandeja.
        Verifica que el tama√±o y contenido de la nueva bandeja sean correctos.
    
    """
    
    def __init__(self):
        super().__init__()
        self._bandeja = []

    @property
    def bandeja(self):
        return self._bandeja
        
    @bandeja.setter
    def bandeja(self, nueva_bandeja):
        if len(nueva_bandeja) > 3:
            raise ValueError('nueva_bandeja tiene m√°s de 3 frutas.')
        for fruta in nueva_bandeja:
            if not isinstance(fruta, dict):
                raise TypeError(f'fruta {fruta} no es diccionario')
            if not all(key in fruta for key in ["nombre", "tipo", "color", "pepas"]):
                raise ValueError('falta una llave en la fruta entregada')
        self._bandeja = nueva_bandeja                

    def listar_ingredientes(self):
        if not self.bandeja:
            print("Bandeja vac√≠a")
        else:
            print(f"Ingredientes en la bandeja: {', '.join((fruit['nombre'] for fruit in self.bandeja))}")
    
    def bandeja_vacia(self):
        if not self.bandeja:
            raise Exception(
                'Error ‚ùå: La bandeja no tiene ingredientes.'
            )
    
    def preparar_jugo(self):
        try:
            self.esta_enchufado()
            self.bandeja_vacia()
            if filtrar_por_pepa(self.bandeja, True):
                print('Alerta ‚ö†Ô∏è: El jugo puede contener restos de pepas.')
            mensaje = f"Jugo de {', '.join((fruit['nombre'] for fruit in self.bandeja))} listo. üèñÔ∏èü•§ Que lo disfrutes!!! ü•§üèñÔ∏è. "
            self.bandeja = []
        except Exception as e:
            mensaje = e
        return mensaje
````
